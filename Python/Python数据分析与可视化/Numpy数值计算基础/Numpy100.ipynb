创建数组

```python
import numpy as np
print('通过numpy创建一维数组')
arr=np.array([1, 2, 3])
print(arr)
print('通过numpy创建二维数组')
arr=np.array([(1, 2, 3), (4, 5, 6)])
print(arr)
print('通过numpy创建全为0的数组')
arr=np.zeros((3, 3))
print(arr)
print('通过numpy创建全为1的数组')
arr=np.ones((2, 3, 4))
print(arr)
print('创建一维等差数组')
arr=np.arange(5)
print(arr)
print('创建二维等差数组')
arr=np.arange(6).reshape(2, 3)
print(arr)
print('创建单位矩阵')
arr=np.eye(3)
print(arr)
print('创建等间隔一维数组')
arr=np.linspace(1, 10, num=6)
print(arr)
print('创建二维随机数组')
arr=np.random.rand(2,3)
print(arr)
print('创建二维随机整数数组（数值小于 5）')
arr=np.random.randint(5, size=(2, 3))
print(arr)
print('依据自定义函数创建数组')
arr=np.fromfunction(lambda i, j: i + j, (3, 3))
print(arr)
```

数组运算

```python
print('生成一维示例数组')
a = np.array([10, 20, 30, 40, 50])
b = np.arange(1, 6)
print('a=',a)
print('b=',b)
print('一维数组加法运算')
print(a + b)
print('一维数组减法运算')
print(a - b)
print('一维数组乘法运算')
print(a * b)
print('一维数组除法运算')
print(a / b)
print('生成二维示例数组（可以看作矩阵）')
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
print('A=',A)
print('B=',B)
print('矩阵加法运算')
print(A + B)
print('矩阵减法运算')
print(A - B)
print('矩阵元素间乘法运算')
print(A * B)
print('矩阵乘法运算')
print(np.dot(A,B))
print('如果使用 np.mat 将二维数组准确定义为矩阵，就可以直接使用 * 完成矩阵乘法计算')
print(np.mat(A) * np.mat(B))
print('数乘矩阵')
print(2 * A)
print('矩阵的转置')
print(A.T)
print('矩阵求逆')
print(np.linalg.inv(A))
```
